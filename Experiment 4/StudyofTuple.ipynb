```python
'''An organization records the attendance of all employees in a particular week'''


employee_attendance = ("P", "P", "A", "P", "P", "A", "P")
print(f"Employee's weekly attendance record: {employee_attendance}")


present_days = employee_attendance.count("P")
print(f"Total number of present days: {present_days}")


absent_days = employee_attendance.count("A")
print(f"Total number of absent days: {absent_days}")


if "A" in employee_attendance:
    print("The employee was absent at least once during the recorded period.")
else:
    print("The employee was present every day during the recorded period.")
```
