```python
## Access Items
football = ["Ronaldo" , "Messi" , "Neymar"]
print(football[1])
```

```python
## Reverse/Negative Indexing
football = ["Ronaldo" , "Messi" , "Neymar" , " Zidane" , "Ronaldinho" , "Kaka"]
print(football[-3])
print(football[-3:])
```

```python
## Slicing
football = ["Ronaldo" , "Messi" , "Neymar", "Zidane" , "Ronaldinho" , "Kaka"]
print(football[2:5])
```

```python
## Slicing
football = ["Ronaldo" , "Messi" , "Neymar", "Zidane" , "Ronaldinho" , "Kaka"]
print(football[:4])
```

```python
## Slicing
football = ["Ronaldo" , "Messi" , "Neymar", "Zidane" , "Ronaldinho" , "Kaka"]
print(football[2:])
```

```python
## Change Item Value
football = ["Ronaldo" , "Messi" , "Neymar"]
football[1] = "Mbappe"
print(football)
```

```python
## Append Items
football = ["Ronaldo" , "Messi" , "Neymar"]
football.append("Zidane")
print(football)
```

```python
## Insert Items
football = ["Ronaldo" , "Messi" , "Neymar"]
football.insert(1 , "Zidane")
print(football)
```

```python
## Extend List
football = ["Ronaldo" , "Messi" , "Neymar"]
football1 = ["Zidane" , "Ronaldinho" , "Kaka"]
football.extend(football1)
print(football)
```

```python
marks = [78 , 85 , 90 , 66 , 88]
print("Maximum Marks is:" , max(marks))
print("Minimum Marks is:" , min(marks))
print("Total Marks is:" , sum(marks))
print("Number of students is:" , len(marks))
print("Average Marks is:" , sum(marks)/len(marks))
print("Sort the marks in ascending order" , marks.sort())
print(marks)
```

```python
'''2. Grocery Shopping List
A user maintains a grocery list for monthly shopping.'''
grocery_list = ["Milk" , "Egg" , "Butter" , "Bread" , "Chicken"]
print("Created a grocery list:" , grocery_list)
grocery_list.append("Rice")
print("Added new items to the list:" , grocery_list)
grocery_list.remove("Egg")
print("Removed purchased items;" , grocery_list)
print("Final shopping list:" , (grocery_list))
```

```python
'''3. A person stores phone contacts in a list.'''
phone_contacts = ["Virat" , "Dhoni" , "Rohit" , "Shreyas"]
print("Created a new list:" , phone_contacts)
phone_contacts.append("Sachin")
print("Added a new contact:" , phone_contacts)
phone_contacts.remove("Rohit")
print("Deleted an existing contact:" , phone_contacts)
phone_contacts.sort()
print("Displayed contacts alphabetically:" , phone_contacts)
```

```python
'''4. Temperature Analysis.'''
temperature_analysis = [18 , 20 , 22 , 24 ,26 , 28 , 30 , 32 , 34 , 36]
print("Temperature of first 5 days:" , temperature_analysis[:5])
print("Temperature of last 5 days:" , temperature_analysis[-5:])
print("Maximum temperature:" , max(temperature_analysis))
print("Minimum temperature:" , min(temperature_analysis))
print("Average temperature:" , sum(temperature_analysis)/len(temperature_analysis))
```

```python
'''5. Daily Expense Tracker.'''
daily_expense_tracker = [1000 , 2000 , 3000 , 4000 , 5000 , 6000 , 7000 , 8000 , 9000 , 10000]
print("Total Expense:" , sum(daily_expense_tracker))
print("Average Expense:" , sum(daily_expense_tracker)/len(daily_expense_tracker))
print("Highest Spending Day:" , max(daily_expense_tracker))
print("Lowest Spending Day:" , min(daily_expense_tracker))
```
