```python
# Preprocessing on a Dataset - 1
import pandas as pd

data = {
    "Roll no":[101,102,103,104,105],
    "Name" : ["Clark","Lana","Lex","Sydney","Lois"],
    "Age" : [20, "-", 21, 20, "22"],
    "Marks" : [85,90, "-", 88,76],
    "Department" : ["Aero","IT","Automobile","ECE","CSE"],
    "Admission Date" : ["01-06-2023", "03-06-2023","05-06-2023","07-06-2023","12-06-2023"]
}

df = pd.DataFrame(data)
df.to_csv("student_data.csv", index=False)
```

```python
import pandas as pd
import numpy as np

# Step 1: Load the dataset
df = pd.read_csv("student_data.csv")

print("Original Data: ")
print(df)
```

```python
# Step 2: Replace '-' with NaN to mark missing vlaues

df.replace("-", np.nan, inplace=True)

#Convert columns to numeric where needed

df["Age"] = pd.to_numeric(df["Age"], errors = 'coerce')
df["Marks"] = pd.to_numeric(df["Marks"], errors = 'coerce')

# When converting data (like strings to datetime) if some values are invalid or cannot be converted, instead of throwing an error, it will: Replace thoe invald values with NaT (Not a Time)

df
```

```python
# Step 3: Check missing values

print("Missing Values:")
print(df.isnull().sum())
```

```python
# Fill missing Age with mean

df["Age"].fillna(df["Age"].mean(), inplace=True)
```

```python
print(df.isnull().sum())
```

```python
# Step 5: Fix inconsistent text formatting

df["Department"] = df["Department"].str.upper()
df
```

```python
# Step 6: Convert admission_date to standard datetime format

df["Admission Date"] = pd.to_datetime(df["Admission Date"], errors = 'coerce')
df
```

```python
# Step 7: Show cleaned data

print("Cleaned Data:")
print(df)
```

```python
# Step 8: Save cleaned dataset

df.to_csv("cleaned_student_data.csv", index = False)
print("\n Cleaned CSV file saved as cleaned_student_data.csv")
```

```python
# Working with missing values on Dataset -2

df=pd.read_csv('/content/Cars93.csv')
df
```

```python
df.isna().sum() # Attritbutes are displayed
```

```python
# Imputing missing values

#1. Luggage Room
df['Luggage.room'] = df['Luggage.room'].fillna(value=df['Luggage.room'].mean())
```

```python
#2. Rear seat room
df['Rear.seat.room'] = df['Rear.seat.room'].fillna(value=df['Rear.seat.room'].mean())
```

```python
#3. Airbags
df['AirBags'] = df['AirBags'].fillna(value=df['AirBags'].mode().values[0])
```

```python
# Save cleaned dataset
df.to_csv("cleaned_Cars93.csv", index= False)
```
