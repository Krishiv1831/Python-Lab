# Sets

```python
# Creation of set
football = {"Ronaldo", "Messi", "Neymar", "Mbappe"}
print(football)
```

```python
# Duplicates not allowed
football = {"Ronaldo", "Messi", "Neymar", "Mbappe", "Ronaldo"}
print(football)
```

```python
# The values True and 1 are considered the same value in the sets, and are treated as duplicates:
football = {"Ronaldo", "Messi", "Neymar", "Mbappe", True, 1}
print(football)
```

```python
# A set with strings, integers and boolean values :
set1 = {"xyz", 18, True, 31, "female"}
print(set1)
```

```python
# To check membership
football = {"Ronaldo", "Messi", "Neymar", "Mbappe"}
print("Ronaldo" in football)
print("Lewandowski"  in football)
```

```python
# Once a set is created, you cannotreplace its items,
# but you can add new items.
football = {"Ronaldo", "Messi", "Neymar", "Mbappe"}
football.add("Lewandowski")
football.remove("Messi")
print(football)
```

```python
# Set Operations
X = {1, 3, 5, 7}
Y = {2, 4, 6, 8}

print("Union:" , X | Y)
print("Intersection:" , X & Y)
print("Difference:" , X - Y)
print("Symmetric Difference:" , X ^ Y)
```

# Python Frozenset

```python
#frozenset
y = frozenset({"Ronaldo", "Messi", "Neymar", "Mbappe" })
print(y)
print(type(y))
```

```python
# Adding or Removal of elements are not allowed in the frozenset
y = frozenset({"Ronaldo", "Messi", "Neymar", "Mbappe" })
y.add("Lewandowski")
y.remove("Messi")
print(y)
```

```python
'''Problem statement 3: One set contains students in the Cricket Club, another in the Fooball Club.
Find students who are : 1. In both clubs 2. In only one club
Concept used: Intersection, Difference'''
cricket = {"Krishiv" , "Harshit" ,"Aditya"  , "Swastik"}
football = {"Harshit" , "Shekhar" , "Swastik" , "Aditya" , "Krishiv"}
print("Students in both clubs:" , cricket & football)
print("Students in only one club:" , cricket ^ football)
```

```python
'''Problem Statement 4: Absent Students
Given a set of all students and a set of present students
find the absent students.
Concept Used: Difference'''
all_students = {"Harshit" , "Shekhar" , "Swastik" , "Aditya" , "Krish" , "Khush" , "Kshitij" , "Vishwam"}
present_students = {"Harshit" , "Shekhar" , "Swastik" , "Khush" , "Vishwam"}
absent_students = all_students - present_students
print("Absent Students:" , absent_students)
```

```python
'''Problem Statement 5: Remove Invalid Entries.
A set contains valid course codes.
Renove a discontinued course codes from the set.
Concept Used: remove() / discard()'''
course_codes = {"CS101", "CS102", "CS103", "CS104", "CS105"}
print("Current Course Codes:", course_codes)
discontinued_course = "CS103" , "CS101"
course_codes.difference_update(discontinued_course)
print("Updated Course Codes:", course_codes)
```

# Dictionaries

```python
#Creating a Dictionary
football = {
    "Player1" : "Ronaldo",
    "Player2" : "Messi",
    "Player3" : "Neymar",
}
football
```

```python
# Dictionary takes latest value given for the key. It will not allow duplicates
football = {
    "Player1" : "Ronaldo",
    "Player2" : "Messi",
    "Player3" : "Neymar",
    "Player2" : "Mbappe",
}
football
```

```python
# Finding the number of members & the type in a dictionary
football = {
    "Player1" : "Ronaldo",
    "Player2" : "Messi",
    "Player3" : "Neymar",
    "Player4" : "Mbappe",
}
print(len(football))
print(type(football))
```

```python
# Changing/Adding key: value in a dictionary
football = {
    "Player1" : "Ronaldo",
    "Player2" : "Messi",
    "Player3" : "Neymar",
    "Player4" : "Mbappe",
}
y = football.values()
print(y) #before the change
football["Player2"] = "Lewandowski"
print(y) #after the change
```

```python
'''2. A dictionary stores student names and marks.
Write a python program to search for a students's marks using the name entered by the user.
If the student is not found, display a suitable message.'''
students = {"Harshit":99 , "Krish":67 , "Kshitij":69 , "Khush":98 , "Swastik":76 , "Vishwam": 47}
name = input("Enter the name of the student: ")
if name in students:
    print(f"The marks of {name} are {students[name]}")
else:
    print(f"{name} is not found in the dictionary.")
```

```python
'''3. A dictionary stores usernames and passwords.
Write a python program to validate user login.'''
login = {"admin": "password123", "user1": "abc@123", "guest": "guestpass"}
username = input("Enter username: ")
password = input("Enter password: ")

if username in login:
    if login[username] == password:
        print(f"Welcome, {username}! Login successful.")
    else:
        print("Incorrect password. Please try again.")
else:
    print("Username not found. Please register or check your username.")
```

```python
'''4. A dictionary stores student names and marks.
Write a Python program to find the student who scored the highest marks.'''
students = {"Harshit":99 , "Krish":67 , "Kshitij":69 , "Khush":98, "Swastik":76 , "Vishwam": 47}
topper = max(students, key=students.get)
print("The student with the highest marks is ", topper)
print("The marks of the topper is", students[topper])
```
