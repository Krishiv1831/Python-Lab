```python
import pandas as pd

data = {
  'Order_ID':['01','02','03','04','05','06','07','08','09','010'],

  'Category':['Electronics','Clothing','Electronics','Grocery','Electronics','Clothing','Grocery','Electronics','Clothing','Electronics'],

  'Payment_Method':['Credit Card','PayPal','Debit Card','PayPal','Credit Card','PayPal','Debit Card','Credit Card','PayPal','Debit Card'],

  'Delivery_Type':['Standard','Express','Standard','Express','Standard','Express','Standard','Express','Standard','Express'],

  'Customer_Type':['Regular','New','Regular','New','Regular','New','Regular','New','Regular','New'],
}

df = pd.DataFrame(data)

print(df)
```

```python
#1. Frequency Count
# Count how many orders in each category.

print(df['Category'].value_counts())
print()
print(df['Payment_Method'].value_counts())
print()
print(df['Delivery_Type'].value_counts())
print()
print(df['Customer_Type'].value_counts())
```

```python
#2. Unique Values
# Display all unique categories.

print(df['Category'].unique())
print()
print(df['Payment_Method'].unique())
print()
print(df['Delivery_Type'].unique())
print()
print(df['Customer_Type'].unique())
```

```python
#3. Number of Unique Values

print(df['Category'].nunique())
print()
print(df['Payment_Method'].nunique())
print()
print(df['Delivery_Type'].nunique())
print()
print(df['Customer_Type'].nunique())
```

```python
#4. Cross Tabulation
# Category vs Payment Method

print(pd.crosstab(df['Category'], df['Payment_Method']))
print()
```

```python
#5. Cross Tabulation (Delivery vs Customer Type)

print(pd.crosstab(df['Delivery_Type'], df['Customer_Type']))
print()
```

```python
#6. Percentage Distribution

print(df['Category'].value_counts(normalize=True) * 100)
print()
```

```python
#7. Filtering Data
# Example: Orders from Electronics category

electronics_orders = df[df['Category'] == 'Electronics']
print(electronics_orders)
```

```python
#8. Grouping
# Orders by category and payment method

print(df.groupby(['Category', 'Payment_Method']).value_counts())
print()
```

```python
#9. Sorting Data

sorted_df = df.sort_values(by='Category')
print(sorted_df)
```

```python
# Second Dataset

import pandas as pd

df = pd.read_csv("/content/Expt11.csv")

print(df.head())
```

```python
#1. Count Students in Each Grade
# Find how many students obtained each grade.

print("Grade Frequency")
print(df['Grade'].value_counts())
```

```python
#2. Percentage Distribution of Grades

print("Percentage Distribution")
print(df['Grade'].value_counts(normalize=True) * 100)
```

```python
#3. Number of students in Each Department

print("Department Frequency")
print(df['Department'].value_counts())
```

```python
#4. Gender Distribution

print("Gender Distribution")
print(df['Gender'].value_counts())
```

```python
#5. Gender vs Grade Analysis

print(pd.crosstab(df['Gender'], df['Grade']))
print()
```

```python
#6. Department vs Gender

print(pd.crosstab(df['Department'], df['Gender']))
print()
```

```python
#7. Department vs Grade Percentage

print(pd.crosstab(df['Department'], df['Grade'], normalize=True) * 100)
```

```python
#8. Grouping Data (Departmentwise)

print(df.groupby('Department').value_counts())
```
