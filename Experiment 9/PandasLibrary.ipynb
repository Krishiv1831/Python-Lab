```python
# Add a New Column
'''If marks 80-90 : Grade A
If marks > 90 : Distinction
If marks > 95-100 : Grade B
If marks < 80 : Fail'''

df["Grade"] = ["Distinction" , "Fail" , "Fail" , "Fail" , "Grade A"]
df
```
